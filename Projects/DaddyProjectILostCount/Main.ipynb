```python
import pandas as pd
```

```python
data = pd.read_csv("data/combine_Data.csv")
data
```

```python
cols = list(data.columns.values)
d2 = pd.DataFrame(columns = cols)
d2
```

```python
d2 = d2.append(data.iloc[0])
```

```python
d2.drop(0, inplace=True)
```

```python
d2
```

```python
for i in range(len(data)):
    match = False
    row1 = list(data.iloc[i])
    for j in range(len(data)):
        if i == j:
            continue
        else:
            row2 = list(data.iloc[j])
            rw2 = data.iloc[j]
            if row1[0] == row2[0] and row1[3] == row2[3] and row1[4] == row2[4] and row1[8] == row2[8]:
                row3 = row1
                met = row1[7] + row2[7]
                row3[7] = met
                d2.loc[i] = row3
                match = True
                break
    if match == False:
        d2.loc[i] = row1
```

```python
d2.drop_duplicates(subset=None, keep='first', inplace=True)
```
