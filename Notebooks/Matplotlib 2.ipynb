```python
#Import
import numpy as np
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
plt.style.use("ggplot")
from numpy.random import randn
import pandas as pd
%matplotlib inline
```

```python
x = np.array(range(-15,14))
y = eval("x**2 + 2*x + 5")
y2 = eval("x**3")
y3 = eval("2*x + 4")
y4 = randn(100)
```

```python
plt.plot(x,y,label='x**2 + 2*x + 5')
plt.plot(x,y2,label='x**3')
plt.plot(x,y3,label='2*x + 4')
x2 = randn(100)
plt.scatter(x2,y4,label="Scatter plot")
axes = plt.gca()
axes.set_xlim([-15,13])
axes.set_ylim([-100,100])
plt.legend()
```
