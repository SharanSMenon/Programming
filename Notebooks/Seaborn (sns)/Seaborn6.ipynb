# Matrix Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
%matplotlib inline
```

```python
flights.head()
```

## Heatmaps

```python
tc = tips.corr()
```

```python
sns.heatmap(tc)
```

```python
tc
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc,annot=True)
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
sns.heatmap(tc,annot=True,cmap='cool')
```

```python
# Working with flights dataset
flights.head()
```

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp,annot=True,cmap='cool',fmt='d')
```

```python
het = sns.heatmap(fp,linewidths=1,annot=True,fmt='d')
```

# Clustermaps

```python
sns.clustermap(fp)
```

```python
sns.clustermap(tc)
```

```python
fig = sns.clustermap(fp,annot=True,fmt='d')
```
