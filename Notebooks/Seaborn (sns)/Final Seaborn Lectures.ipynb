```python
import numpy as np
import pandas as pd
from numpy.random import randn
import matplotlib as mpl
from matplotlib import pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
from scipy import stats
%matplotlib inline
```

## Regression plots

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot('total_bill','tip',tips)
```

```python
sns.lmplot('total_bill','tip',tips,
          scatter_kws={
              'marker':'d',
              'color':'indianred'
          },
          line_kws={
              'linewidth':1,
              'color':'blue',
          })
```

```python
sns.lmplot('total_bill','tip',tips,fit_reg=True,scatter=False)
```

```python
sns.lmplot('total_bill','tip',tips,fit_reg=False)
```

```python
tips['tip_pect'] = 100*(tips['tip']/tips['total_bill'])
```

```python
tips.head()
```

```python
plt.hist(tips['sex'],bins=2,edgecolor='black')
plt.xlabel('Sex')
plt.ylabel('# of people')
```

```python
# sns.barplot(x='sex',y='')
```

```python
sns.lmplot('size','tip_pect',tips)
```

```python
# Jitter
sns.lmplot('size','tip_pect',tips,x_jitter=0.8)
```

```python
sns.lmplot('size','tip_pect',tips,x_estimator=np.mean)
```

```python
sns.lmplot('total_bill','tip_pect',tips,hue='sex',markers=['x','o'])
```

```python
sns.lmplot('total_bill','tip_pect',tips,hue='day')
```

```python
sns.lmplot('total_bill','tip_pect',tips,lowess=True,line_kws={
    'color':'black'
})
```

```python
sns.regplot('total_bill','tip_pect',tips)
```

```python
fig, (axis1,axis2) = plt.subplots(1,2,sharey=True)
sns.regplot('total_bill','tip_pect',tips,ax=axis1)
sns.violinplot(tips['size'],tips['tip_pect'],ax=axis2)
```
