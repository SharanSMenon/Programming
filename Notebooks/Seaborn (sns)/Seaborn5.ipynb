```python
import seaborn as sns
%matplotlib inline
sns.set_style("whitegrid")
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
# Counts the instances
sns.countplot(x='sex',data=tips)
```

```python
import matplotlib.pyplot as plt
```

```python
# Same thing as countplot
plt.hist(tips['sex'],bins=2,edgecolor='black')
```

```python
# Box plots
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
# Violin plots
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
# Compare the 2 plots
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
# Stripplot
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,dodge=True,hue='sex')
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
# Factor plot
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='box')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='swarm')
```
