# Matplotlib from Jose portilla

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y= x ** 2
```

```python
x
```

```python
y
```

```python
# Functional Method
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Functional method')
```

```python
#Many plots on the sames canvas
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# OO method
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("OO method")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,1,0.8])
axes2 = fig.add_axes([0.2,0.5,0.5,0.3])
axes1.plot(x,y)
axes2.plot(y,x,'r')
axes1.set_title("Plot 1")
axes2.set_title("Plot 2")
```

```python
#Subplots with OO method
fig,axes = plt.subplots(nrows=1,ncols=2)
# for ax in axes:
#     ax.plot(x,y)
axes[0].plot(x,y)
axes[0].set_title("First Plot")
axes[1].plot(y,x)
axes[1].set_title("Second Plot")
# plt.tight_layout()
```

```python
axes
```

# Figure Size and DPI

```python
fig,ax = plt.subplots(nrows=2,ncols=1,figsize=(5,4))
# ax = fig.add_axes([0,0,1,1])
ax[0].plot(x,y)
ax[1].plot(y,x,'r')
plt.tight_layout()
```

```python
fig.savefig('myfig1.png',dpi=200)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="X Squared")
ax.plot(x,x**3,label="X Cubed")
ax.legend(loc=(0.75,0.3))
# ax.set_title("Title")
# ax.set_ylabel("Y")
# ax.set_xlabel("X")
```

```python
fig.savefig("legendFig1.png",dpi=200)
```

# Matplotlib part 3
Customization of figure

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=7,ls="-",marker='o',markersize=10,
       markerfacecolor="orange",markeredgecolor="orange",
       markeredgewidth=3) 
# RGB Hex codes
# Transparency
# Line Style
# Line width
```

```python
fig.savefig("customLine.png")
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple') 
ax.set_xlim(0,1)
ax.set_ylim(0,1)
```
